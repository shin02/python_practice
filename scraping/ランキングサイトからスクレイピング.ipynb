```python
from selenium import webdriver
```

```python
browser = webdriver.Chrome()
```

```python
browser.get("https://scraping-for-beginner.herokuapp.com/ranking/")
```

```python
#観光地名の取得
titles = []
reviews = []
itemised_reviews =[]

#ページ数を取得
page_last = int(browser.find_elements_by_class_name("waves-effect")[-2].text)+1

#全ページの情報を取得
for n in range(1,page_last):
    
    #ページの読み込み
    browser.get("https://scraping-for-beginner.herokuapp.com/ranking/?page={}".format(n))

    #ページの情報を取得
    elems_rankingBox = browser.find_elements_by_class_name("u_areaListRankingBox")

    for elem_rankingBox in elems_rankingBox:

        #タイトルを取得
        title = elem_rankingBox.find_element_by_class_name("u_title").text.split("\n")[1]
        titles.append(title)

        #総合評価の取得
        review = elem_rankingBox.find_element_by_class_name("u_rankBox").find_element_by_class_name("evaluateNumber").text
        reviews.append(float(review))

        #各項目ごとの評価の取得
        elems = elem_rankingBox.find_elements_by_class_name("is_rank")

        #観光地nの項目ごとの評価を入れるリストを作成
        itemised_review = []

        for elem in elems:
            itemised_review.append(float(elem.text))

        #観光地nの項目ごとのリストを全体のリストに追加
        itemised_reviews.append(itemised_review)
```

```python
#pandasの形式に変換
import pandas as pd
```

```python
df = pd.DataFrame()

df["title"] = titles
df["review"] = reviews

df_categories = pd.DataFrame(itemised_reviews)
df_categories.columns = ["楽しさ","人混みの多さ","景色","アクセス"]

pd.concat([df,df_categories],axis=1)
```
